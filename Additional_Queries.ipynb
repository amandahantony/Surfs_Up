```python
import numpy as np
import pandas as pd


import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from sqlalchemy import extract
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")


Base = automap_base()
Base.prepare(engine, reflect=True)
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
session = Session(engine)
```

## Calculating Avg Temps in Summer Months - Ice Cream Sales

```python
results = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date ) == 6).all()
```

```python
June_temps_df = pd.DataFrame(results, columns=['Date','Temperature'])
June_temps_df.set_index(June_temps_df['Date'], inplace=True)
June_temps_df = June_temps_df.sort_index()
print(June_temps_df.to_string(index=False))
```

```python
June_avg_temp = June_temps_df[['Temperature']].mean()
June_avg_temp
```
